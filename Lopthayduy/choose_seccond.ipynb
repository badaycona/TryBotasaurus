```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
import randomly_chosen_sample_of_5
import matplotlib.animation as animation


```

```python
results = []  
alpha = 0.05
smaller = [0]
total = [0]
fig, ax = plt.subplots()
_, bins, patches = ax.hist([], bins=20, range=(0,1), color='skyblue', edgecolor='black')
def update_histogram(frame_number):
    for i in range(20):
        total[0] += 1
        new_p_value = randomly_chosen_sample_of_5.gen_p_value()
        if new_p_value <= alpha:
            smaller[0] += 1
            print('Increased')
        results.append(new_p_value)

    current_counts, _ = np.histogram(results, bins=bins)

    for patch, new_height in zip(patches, current_counts):
        patch.set_height(new_height)

    ax.set_ylim(0, max(current_counts) * 1.1 + 5)
    status_text.set_text('%.2f'%(smaller[0] / total[0]))

    return list(patches) 



ax.set_xlim(0, 1)
ax.set_xlabel('P-value')
ax.set_ylabel('Frequency')
iter = 50

status_text = ax.text(0.95, 0.95,'', transform=ax.transAxes,
                      fontsize=10, color='red',
                      horizontalalignment='right', verticalalignment='top')


ani = animation.FuncAnimation(fig, update_histogram, frames=iter,
                              interval=10, blit=True, repeat=False)
```
